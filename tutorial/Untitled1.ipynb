```python
from cij.util import units
units.Unit
```

```python
from typing import NamedTuple, List
class ResultsWriterRule(NamedTuple):

    keywords: List[str]
    fname_pattern: str
    prop: str
    unit: units.Unit
    unit_internal: units.Unit
    var_type: str

    @classmethod
    def create(cls, rule: dict):
        cls(
            rule["keywords"],
            rule["file_name_pattern"],
            rule["type"],
            rule["unit_internal"],
            rule["unit"],
            rule["property"]
        )
```

```python
ResultsWriterRule(var_type="", unit_internal = "", prop="")
```
